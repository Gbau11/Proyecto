```python
df_emiss2 = df_emiss2.drop(['Energy_type','Unnamed: 0','Energy_consumption','Energy_production','Energy_intensity_per_capita','Population','Energy_intensity_by_GDP'], axis=1)

df_emiss2
```

```python
df_emis_filtrado = df_emiss2.loc[~df_emiss2['Year'].isin([1992, 1991, 1990,1989,1988,1987, 1986, 1985,1984,1983, 1982,1981,1980])]

df_emis_filtrado
```

```python
df_emis_filtrado1 = df_emis_filtrado.drop('CO2_emission',axis=1)
df_emis_filtrado2 = df_emis_filtrado.drop('GDP',axis=1)
df_emis_filtrado1
```

```python
df_emis_filtrado1_pivot = df_emis_filtrado1.pivot_table(index='Country', columns='Year', values='GDP', aggfunc='first')
df_emis_filtrado1_pivot
```
